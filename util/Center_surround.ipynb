```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.convolution import MexicanHat2DKernel
```

```python
gabor = np.load("/home/ziniuw/Tangdata/gabor.npy")
gabor = gabor.reshape(24,1,10,10)
res = np.zeros((27,1,10,10))
res[0:24] = gabor
```

```python
j = 24
for i in [2,4,8]:
    mexicanhat_2D_kernel = MexicanHat2DKernel(i, x_size=10, y_size=10)
    plt.imshow(mexicanhat_2D_kernel, interpolation='none', origin='lower', cmap='gray')
    plt.xlabel('x [pixels]')
    plt.ylabel('y [pixels]')
    plt.show()
    y = mexicanhat_2D_kernel._array * 1e5
    y = y/np.max(y)*255
    
    res[j] = y
    j+=1
```
